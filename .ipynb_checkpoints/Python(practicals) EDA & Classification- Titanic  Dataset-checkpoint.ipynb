#### Handling Missing Values

```python
#we shall drop cabin column as it has most missing value at 78%
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
df.Name.head(50)
df['passenger_title'] = df.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
#As for Age and  Embarked columns we shall replace missing values with the median and  mode  of their column values
df['Age']=df['Age'].fillna(df.Age.median())
#The mode is the most frequently occurring element in a series.
df['Embarked']=df['Embarked'].fillna(df.Embarked.mode()[0])
```

```python
#checking missing values in our dataset
df.isna().sum()
```

This means replacing the column values with the a mode value  Embarked of the passengers onboard and for age replace missing values with their ranging age values from the Pclasses.
Now our data has no missing values.

```python
#drop passengerId column as it has no correlation with the outcome in whether a passenger survived or not.
df.drop('PassengerId',axis=1,inplace=True)

#we can all drop the name of the passenger as we are mostly interested in the gender to know the survival classification
#df.drop('Name',axis=1,inplace=True)
```

```python
df.head()
```

#### Setting and repositioning  our Target variable

```python
#relocating our target column(Survived) to be the last column.
cols=list(df.columns.values) #create a list of our columns
cols.pop(cols.index('Survived'))#removing Survived from our column list
df=df[cols+['Survived']]#create a new list add the survival column that will added as the last column
```

It's a good practice to have the target varible column as the last column.
From our target varible column it consists of numerical values in binary from 1 an 0.

#### Creating feature and outcome data variables from our dataset

```python
df['norm_fare'] = np.log(df.Fare+1)
df['norm_fare'].hist()
```

```python
#Creating our feature and outcome data variables
X=df.drop(['Name','passenger_title','Ticket','Survived'],axis=1)
y=df.Survived
```

I have dropped the ticket column will be looking if it has an impact in the overall model performance.

##### Covert categorical column values  into number values

```python
from sklearn.preprocessing import LabelEncoder
lbl_enc=LabelEncoder()
#transforming categofircal values into numerical values for our model to use
X['Sex']=lbl_enc.fit_transform(X['Sex'])
X['Embarked']=lbl_enc.fit_transform(X['Embarked'])
```

```python
X
```

### Model Building.

#### 1)Logistic Regression

```python
#defining our model
log_reg=LogisticRegression()
```

```python
#splitting our data into training and testing dataset
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#tuning our model using training dataset
log_reg.fit(train_X,train_y)
```

```python
#using our model to predict passenger surival based on test and train data
pred=log_reg.predict(test_X)
train=log_reg.predict(train_X)
```

```python
#valuating our model performance in predicting passenger survial
round(accuracy_score(test_y,pred),4)
```

#### 2)Decision Tree Classifier

```python
#define our model
dtc_model=DecisionTreeClassifier()
```

```python
#splitting our data into training and testing dataset
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=1)

#tuning our model using training dataset
dtc_model.fit(train_X,train_y)

#using our decision tree classifier model to predict passenger surival based on test and train data
pred=dtc_model.predict(test_X)
train=dtc_model.predict(train_X)

#valuating our decision tree classifier model performance in predicting passenger survial
round(accuracy_score(test_y,pred),2)
```

```python
round(accuracy_score(train_y,train),4)
```

#### 3)Random Forest Classifier

```python
#define our model
rf_model=RandomForestClassifier(n_estimators=100)
```

```python
#splitting our data into training and testing dataset
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=1)

#tuning our random forest  model using training dataset
rf_model.fit(train_X,train_y)

#model prediction using training and testing data 
pred=rf_model.predict(test_X)
train=rf_model.predict(train_X)


#evaluating our Support Vector Machine model performance whien it comes to diagnosis
round(accuracy_score(test_y,pred),4)
```

 #### 4)Support Vector Classifier

```python
#define our model
svm_model=SVC(kernel='rbf')
```

```python
#splitting our data into training and testing dataset
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=1)

#tuning our support vector machine  model using training dataset
svm_model.fit(train_X,train_y)

#model prediction using training and testing data 
pred=svm_model.predict(test_X)
train=svm_model.predict(train_X)


#evaluating our Support Vector Machine model performance whien it comes to diagnosis
round(accuracy_score(test_y,pred),4)
```
